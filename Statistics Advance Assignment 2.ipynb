#### 1. What is hypothesis testing in statistics

Hypothesis testing is a statistical method used to make decisions or inferences about a population based on sample data. It helps determine whether there is enough evidence in a sample to support or reject a certain claim (hypothesis) about a population parameter

#### 2 What is the null hypothesis, and how does it differ from the alternative hypothesis

Null Hypothesis:  A statement in statistics that assumes no effect, no difference, or no relationship between variables in a population. It serves as the default or baseline assumption in hypothesis testing.

Alternative Hypothesis : A statement that contradicts the null hypothesis, suggesting that there is an effect, difference, or relationship in the population. It represents what the researcher aims to prove.

#### 3 What is the significance level in hypothesis testing, and why is it important

The significance level  in hypothesis testing is the probability of rejecting the null hypothesis  when it is actually true. It represents the maximum acceptable risk of making a Type I error (false positive)

#### 4 What does a P-value represent in hypothesis testing 

The P-value (probability value) is the probability of obtaining a test statistic as extreme as or more extreme than the observed result, assuming that the null hypothesis ( is true.

#### 5  How do you interpret the P-value in hypothesis testing

The P-value helps determine whether to reject the null hypothesis  It represents the probability of obtaining the observed results (or more extreme ones) assuming is true.

#### 6 What are Type 1 and Type 2 errors in hypothesis testing

Type 1 Error (False Positive): Rejecting the null hypothesis when it is actually true.
Example:
Claiming a new drug works when it actually doesn't.

Type 2 Error (False Negative): Failing to reject the null hypothesis when it is actually false.
Example:
Failing to detect that a new drug actually works when it does.


#### 7 What is the difference between a one-tailed and a two-tailed test in hypothesis testing

In hypothesis testing, the direction of the alternative hypothesis determines whether we use a one-tailed or a two-tailed test.

One-Tailed Test
Definition: A test where the alternative hypothesis is directional, specifying that the population parameter is either greater than or less than a certain value.

Two-Tailed Test
Definition: A test where the alternative hypothesis is non-directional, stating that the population parameter is different from a certain value, without specifying whether it's higher or lower.

#### 8 What is the Z-test, and when is it used in hypothesis testing

A Z-test is a statistical test used to determine if there is a significant difference between a sample mean and a population mean (or between two sample means) when the population variance is known or the sample size is large enough

#### 9  How do you calculate the Z-score, and what does it represent in hypothesis testing

The Z-score is a standardized score that represents how many standard deviations a data point (or sample mean) is from the population mean. It helps to determine how unusual or extreme a sample result is compared to the population.

#### 10 What is the T-distribution, and when should it be used instead of the normal distribution

The T-distribution (also called Student's t-distribution) is a probability distribution used in hypothesis testing when the sample size is small (ùëõ<30) and the population standard deviation is unknown. It has wider tails than the normal distribution, which helps account for the increased uncertainty with smaller samples.

As the sample size increases, the T-distribution becomes closer to the normal distribution.

#### 11 What is the difference between a Z-test and a T-test

The Z-test and T-test are both statistical tests used to determine if there is a significant difference between a sample and a population, or between two samples. However, they are used in different situations based on the sample size and knowledge of the population's characteristics.

#### 12 What is the T-test, and how is it used in hypothesis testing

The T-test is a statistical test used to compare the means of one or two groups to determine if there is a significant difference between them. It is typically used when:

The sample size is small (ùëõ<30).
The population standard deviation is unknown, and the sample standard deviation is used instead.

#### 13 What is the relationship between Z-test and T-test in hypothesis testing

The Z-test and T-test are both statistical tests used to compare sample data with population data (or between two sample groups) to determine if there is a significant difference. The relationship between the two tests primarily depends on the sample size and whether the population standard deviation is known or unknown

#### 14 What is a confidence interval, and how is it used to interpret statistical results

A confidence interval is a range of values used to estimate the true population parameter (such as a population mean or proportion) based on sample data. It provides an interval estimate of the parameter, along with an associated confidence level that reflects the degree of certainty we have in the estimate.

#### 15 What is the margin of error, and how does it affect the confidence interval

The margin of error is the amount added and subtracted from the sample estimate (such as the sample mean) to create the confidence interval. It accounts for the variability or uncertainty in the estimate, and it defines the range within which we expect the true population parameter to fall.

#### 16 How is Bayes' Theorem used in statistics, and what is its significance

Bayes' Theorem is a fundamental concept in probability theory and statistics that describes how to update the probability of a hypothesis based on new evidence or data. It is used to calculate conditional probabilities, which is the probability of an event occurring given that another event has already occurred.

#### 17 What is the Chi-square distribution, and when is it used

The Chi-square distribution (ùúí2) is a probability distribution that is widely used in statistical inference. It is specifically useful when dealing with categorical data and is a special case of the gamma distribution.

The Chi-square distribution is characterized by a single parameter: the degrees of freedom (df), which is related to the number of independent variables or categories in the data.

#### 18 What is the Chi-square goodness of fit test, and how is it applied

The Chi-square goodness of fit test is a statistical test used to determine whether an observed frequency distribution of a categorical variable matches an expected frequency distribution. This test is useful for comparing observed data to a specific theoretical distribution to see if the observed data "fits" the expected model.

#### 19 What is the F-distribution, and when is it used in hypothesis testing

The F-distribution is a probability distribution used primarily in analysis of variance (ANOVA), regression analysis, and in tests comparing the variances of two or more populations. It is a continuous probability distribution that arises when comparing the ratio of variances from two independent samples.

#### 20 What is an ANOVA test, and what are its assumptions

The ANOVA (Analysis of Variance) test is a statistical technique used to compare the means of three or more groups to determine if at least one group mean is significantly different from the others. It works by analyzing the variance (spread) within each group and between the groups.

The key idea is that if the variance between the groups is significantly greater than the variance within the groups, it suggests that the group means are different from each other.

#### 21 What are the different types of ANOVA tests

ANOVA (Analysis of Variance) is a powerful statistical tool used to compare the means of three or more groups. There are different types of ANOVA tests, each suited to different experimental designs and data structures. The most common types of ANOVA tests are:

One-Way ANOVA
Definition:
One-Way ANOVA is used when you have one independent variable (factor) with three or more levels (groups) and you want to test if there is a significant difference between the means of these groups


Two-Way ANOVA
Definition:
Two-Way ANOVA is used when you have two independent variables (factors), and you want to test their individual and combined effects on the dependent variable. This test also allows you to examine if there is any interaction between the two factors.

#### 22 What is the F-test, and how does it relate to hypothesis testing

The F-test is a statistical test used to compare the variances of two or more groups and assess if they are significantly different. The test statistic follows an F-distribution, and the test is primarily used in the following contexts:

Comparing Variances: To test if the variances of two populations are equal.
Analysis of Variance (ANOVA): To compare the means of multiple groups by analyzing the variance within and between groups.
Regression Analysis: To assess the overall significance of a regression model, particularly whether the predictors explain the variability in the dependent variable.

#### 1 Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results

```python
import numpy as np
from scipy import stats

# Sample data
sample_data = [50, 55, 60, 52, 48, 53, 54, 58, 59, 57]  # Sample data
sample_mean = np.mean(sample_data)  # Sample mean
sample_size = len(sample_data)  # Sample size
population_mean = 55  # Known population mean
population_std_dev = 5  # Known population standard deviation

# Calculate the Z-statistic
z_statistic = (sample_mean - population_mean) / (population_std_dev / np.sqrt(sample_size))

# Calculate the p-value for two-tailed test
p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

# Print results
print(f"Sample Mean: {sample_mean}")
print(f"Z-Statistic: {z_statistic}")
print(f"P-value: {p_value}")

# Interpretation
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: The sample mean is not significantly different from the population mean.")
```

#### 4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Step 1: Simulate random data (sample data)
np.random.seed(42)  # For reproducibility
sample_size = 50
population_mean = 100  # Known population mean
population_std_dev = 15  # Known population standard deviation

# Simulate a sample of data from a normal distribution
sample_data = np.random.normal(loc=population_mean, scale=population_std_dev, size=sample_size)

# Step 2: Set up the null hypothesis
# Null hypothesis (H‚ÇÄ): The sample mean is equal to the population mean (mu = 100)

# Step 3: Calculate the sample mean
sample_mean = np.mean(sample_data)

# Step 4: Perform a two-tailed Z-test
z_statistic = (sample_mean - population_mean) / (population_std_dev / np.sqrt(sample_size))

# Step 5: Calculate the p-value for two-tailed test
p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

# Step 6: Print the results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-statistic: {z_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Step 7: Interpret the results
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: The sample mean is not significantly different from the population mean.")

# Step 8: Visualize the Z-distribution and decision region
# Plot Z-distribution
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x, loc=0, scale=1)

plt.plot(x, y, label='Z-distribution', color='blue')

# Critical values for a two-tailed test at alpha = 0.05
z_critical = stats.norm.ppf(1 - alpha / 2)  # Z-critical value for two-tailed test

# Plot the decision region (critical regions)
plt.fill_between(x, y, where=(x < -z_critical) | (x > z_critical), color='red', alpha=0.5, label="Reject H‚ÇÄ region")

# Plot the calculated Z-statistic
plt.axvline(z_statistic, color='green', linestyle='--', label=f"Z-statistic = {z_statistic:.4f}")

# Add labels and title
plt.title('Two-Tailed Z-Test with Decision Region')
plt.xlabel('Z-value')
plt.ylabel('Density')
plt.legend(loc='best')

# Show the plot
plt.show()
```

#### 5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Function to calculate and visualize Type 1 and Type 2 errors
def visualize_type_errors(population_mean, population_std_dev, sample_size, alpha, true_effect_size):
    # Create normal distributions for H0 and H1
    x = np.linspace(population_mean - 4*population_std_dev, population_mean + 4*population_std_dev, 1000)
    
    # Null hypothesis distribution (H‚ÇÄ) - population mean = population_mean
    h0_dist = norm.pdf(x, population_mean, population_std_dev / np.sqrt(sample_size))
    
    # Alternative hypothesis distribution (H‚ÇÅ) - population mean is shifted by true effect size
    h1_dist = norm.pdf(x, population_mean + true_effect_size, population_std_dev / np.sqrt(sample_size))
    
    # Critical value (Z-critical) for two-tailed test
    z_critical = norm.ppf(1 - alpha / 2)
    
    # Calculate Type 1 Error (area under H‚ÇÄ distribution in rejection region)
    type_1_error = 1 - norm.cdf(z_critical)
    
    # Calculate Type 2 Error (area under H‚ÇÅ distribution in non-rejection region)
    type_2_error = norm.cdf(z_critical) - norm.cdf(-z_critical)
    
    # Plot the distributions
    plt.figure(figsize=(10, 6))
    
    # Plot H‚ÇÄ and H‚ÇÅ distributions
    plt.plot(x, h0_dist, label='H‚ÇÄ Distribution', color='blue')
    plt.plot(x, h1_dist, label='H‚ÇÅ Distribution (True Effect)', color='red')
    
    # Fill the rejection regions (Type 1 error area)
    plt.fill_between(x, 0, h0_dist, where=(x < -z_critical), color='green', alpha=0.3, label='Type 1 Error Region')
    plt.fill_between(x, 0, h0_dist, where=(x > z_critical), color='green', alpha=0.3)
    
    # Fill the Type 2 error region
    plt.fill_between(x, 0, h1_dist, where=(x < z_critical) & (x > -z_critical), color='orange', alpha=0.3, label='Type 2 Error Region')
    
    # Add labels and legend
    plt.title('Visualization of Type 1 and Type 2 Errors')
    plt.xlabel('Sample Mean')
    plt.ylabel('Density')
    plt.legend(loc='upper left')
    
    # Show plot
    plt.show()
    
    # Print Type 1 and Type 2 error values
    print(f"Type 1 Error (false positive): {type_1_error:.4f}")
    print(f"Type 2 Error (false negative): {type_2_error:.4f}")

# Example Usage
population_mean = 100
population_std_dev = 15
sample_size = 30
alpha = 0.05  # Significance level
true_effect_size = 5  # True effect size (shift in population mean for H‚ÇÅ)

# Call the function to visualize the errors
visualize_type_errors(population_mean, population_std_dev, sample_size, alpha, true_effect_size)
```

#### 9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance

```python
import numpy as np
from scipy import stats

def calculate_confidence_interval(sample_data, confidence_level=0.95):
    # Step 1: Calculate the sample mean and sample standard deviation
    sample_mean = np.mean(sample_data)
    sample_std_dev = np.std(sample_data, ddof=1)  # Sample standard deviation
    sample_size = len(sample_data)
    
    # Step 2: Calculate the degrees of freedom
    degrees_of_freedom = sample_size - 1
    
    # Step 3: Find the critical value for Z or T
    if sample_size > 30:  # Use Z-distribution for large samples
        critical_value = stats.norm.ppf(1 - (1 - confidence_level) / 2)
        margin_of_error = critical_value * (sample_std_dev / np.sqrt(sample_size))
    else:  # Use T-distribution for smaller samples
        critical_value = stats.t.ppf(1 - (1 - confidence_level) / 2, degrees_of_freedom)
        margin_of_error = critical_value * (sample_std_dev / np.sqrt(sample_size))
    
    # Step 4: Calculate the confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error
    
    # Step 5: Return the confidence interval
    return (lower_bound, upper_bound)

# Example usage
np.random.seed(42)  # For reproducibility
sample_data = np.random.normal(loc=100, scale=15, size=50)  # Simulated sample data

confidence_level = 0.95  # 95% Confidence Level
ci = calculate_confidence_interval(sample_data, confidence_level)

print(f"Sample Mean: {np.mean(sample_data):.2f}")
print(f"Confidence Interval ({confidence_level*100}%): ({ci[0]:.2f}, {ci[1]:.2f})")
```

#### 10. Write a Python program to calculate the margin of error for a given confidence level using sample data

```python
import numpy as np
from scipy import stats

def calculate_margin_of_error(sample_data, confidence_level=0.95, population_std_dev=None):
    # Step 1: Calculate the sample mean and sample standard deviation
    sample_mean = np.mean(sample_data)
    sample_std_dev = np.std(sample_data, ddof=1)  # Sample standard deviation (use ddof=1 for unbiased estimate)
    sample_size = len(sample_data)
    
    # Step 2: Determine whether to use Z-distribution or T-distribution
    degrees_of_freedom = sample_size - 1
    
    # If population standard deviation is provided, use Z-distribution
    if population_std_dev is not None:
        # Z-critical value for the confidence level
        critical_value = stats.norm.ppf(1 - (1 - confidence_level) / 2)
        margin_of_error = critical_value * (population_std_dev / np.sqrt(sample_size))
    else:
        # T-critical value for the confidence level and degrees of freedom
        critical_value = stats.t.ppf(1 - (1 - confidence_level) / 2, degrees_of_freedom)
        margin_of_error = critical_value * (sample_std_dev / np.sqrt(sample_size))
    
    # Step 3: Return the margin of error
    return margin_of_error

# Example usage:
np.random.seed(42)  # For reproducibility
sample_data = np.random.normal(loc=100, scale=15, size=50)  # Simulated sample data

# Case 1: Known population standard deviation (use Z-distribution)
population_std_dev = 15  # Known population standard deviation
confidence_level = 0.95  # 95% Confidence Level
margin_of_error_z = calculate_margin_of_error(sample_data, confidence_level, population_std_dev)

# Case 2: Unknown population standard deviation (use T-distribution)
margin_of_error_t = calculate_margin_of_error(sample_data, confidence_level)

# Print results
print(f"Sample Mean: {np.mean(sample_data):.2f}")
print(f"Margin of Error using Z-distribution: {margin_of_error_z:.2f}")
print(f"Margin of Error using T-distribution: {margin_of_error_t:.2f}")
```

#### 11.  Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the prior probabilities (P(H))
# Hypothesis H1: The coin is fair (50% heads, 50% tails)
# Hypothesis H2: The coin is biased (80% heads, 20% tails)
prior_H1 = 0.5  # Prior probability of H1 (fair coin)
prior_H2 = 0.5  # Prior probability of H2 (biased coin)

# Define the likelihood (P(D|H)) for each hypothesis based on observed data
# Observed data: 8 heads and 2 tails in 10 flips
observed_heads = 8
observed_tails = 2
n_flips = observed_heads + observed_tails

# Likelihood of observing the data given the hypothesis:
# P(D|H1) = probability of getting 8 heads out of 10 flips with a fair coin
likelihood_H1 = (0.5)**observed_heads * (0.5)**observed_tails
# P(D|H2) = probability of getting 8 heads out of 10 flips with a biased coin (80% heads)
likelihood_H2 = (0.8)**observed_heads * (0.2)**observed_tails

# Step 1: Calculate the marginal likelihood (P(D)) using the law of total probability:
# P(D) = P(D|H1)*P(H1) + P(D|H2)*P(H2)
marginal_likelihood = (likelihood_H1 * prior_H1) + (likelihood_H2 * prior_H2)

# Step 2: Calculate the posterior probabilities (P(H|D)) using Bayes' Theorem:
posterior_H1 = (likelihood_H1 * prior_H1) / marginal_likelihood
posterior_H2 = (likelihood_H2 * prior_H2) / marginal_likelihood

# Print the results
print(f"Posterior Probability of H1 (fair coin): {posterior_H1:.4f}")
print(f"Posterior Probability of H2 (biased coin): {posterior_H2:.4f}")

# Visualize the prior and posterior distributions
hypotheses = ['Fair Coin', 'Biased Coin']
prior_probs = [prior_H1, prior_H2]
posterior_probs = [posterior_H1, posterior_H2]

# Plotting
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

# Prior probability bar plot
ax[0].bar(hypotheses, prior_probs, color=['blue', 'green'])
ax[0].set_title('Prior Probabilities')
ax[0].set_ylabel('Probability')

# Posterior probability bar plot
ax[1].bar(hypotheses, posterior_probs, color=['blue', 'green'])
ax[1].set_title('Posterior Probabilities')
ax[1].set_ylabel('Probability')

plt.tight_layout()
plt.show()
```

#### 13 Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data

```python
import numpy as np
import pandas as pd

def calculate_expected_frequencies(observed_data):
    # Step 1: Create a contingency table from observed data
    contingency_table = np.array(observed_data)
    
    # Step 2: Calculate row totals, column totals, and grand total
    row_totals = contingency_table.sum(axis=1)  # Sum of each row
    column_totals = contingency_table.sum(axis=0)  # Sum of each column
    grand_total = contingency_table.sum()  # Total sum of all data
    
    # Step 3: Calculate expected frequencies using the formula
    expected_frequencies = np.outer(row_totals, column_totals) / grand_total
    
    # Step 4: Return the expected frequencies
    return expected_frequencies

# Example usage:
# Simulated observed data (contingency table)
observed_data = [
    [10, 20, 30],  # Row 1 (e.g., Group 1)
    [6, 9, 12],    # Row 2 (e.g., Group 2)
    [4, 6, 8]      # Row 3 (e.g., Group 3)
]

# Calculate expected frequencies
expected_freqs = calculate_expected_frequencies(observed_data)

# Print the observed data and expected frequencies
print("Observed Data (Contingency Table):")
print(np.array(observed_data))
print("\nExpected Frequencies:")
print(expected_freqs)

```

#### 16  Implement an F-test using Python to compare the variances of two random samples

```python
import numpy as np
from scipy.stats import f

# Step 1: Simulate two random samples
np.random.seed(42)  # For reproducibility

# Sample 1 (Group 1) with variance 25
sample1 = np.random.normal(loc=50, scale=5, size=30)

# Sample 2 (Group 2) with variance 30
sample2 = np.random.normal(loc=50, scale=5.5, size=30)

# Step 2: Calculate the sample variances
var_sample1 = np.var(sample1, ddof=1)  # Sample variance (ddof=1 for unbiased estimate)
var_sample2 = np.var(sample2, ddof=1)  # Sample variance

# Step 3: Calculate the F-statistic (ratio of the variances)
f_statistic = var_sample1 / var_sample2

# Step 4: Calculate degrees of freedom for both samples
df1 = len(sample1) - 1  # Degrees of freedom for sample 1
df2 = len(sample2) - 1  # Degrees of freedom for sample 2

# Step 5: Calculate the p-value (two-tailed test)
p_value = 1 - f.cdf(f_statistic, df1, df2)

# Step 6: Print the results
print(f"Sample 1 Variance: {var_sample1:.4f}")
print(f"Sample 2 Variance: {var_sample2:.4f}")
print(f"F-statistic: {f_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Step 7: Interpret the results
alpha = 0.05  # Significance level
if p_value < alpha:
    print("\nReject the null hypothesis: The variances of the two samples are significantly different.")
else:
    print("\nFail to reject the null hypothesis: The variances of the two samples are not significantly different.")
```

#### 19 Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# Function to check assumptions for ANOVA: normality, independence, and equal variance
def check_anova_assumptions(groups):
    # Assumption 1: Normality Check - Shapiro-Wilk Test
    print("Assumption 1: Normality Check (Shapiro-Wilk Test)")
    for i, group in enumerate(groups):
        stat, p_value = stats.shapiro(group)
        print(f"Group {i+1}: Shapiro-Wilk Test statistic = {stat:.4f}, p-value = {p_value:.4f}")
        if p_value < 0.05:
            print(f"Group {i+1} is not normally distributed (p-value < 0.05).")
        else:
            print(f"Group {i+1} is normally distributed (p-value >= 0.05).")
        print()

    # Assumption 2: Independence Check
    print("Assumption 2: Independence - Assumed based on study design.")
    print("We assume that the samples are independent. This cannot be directly tested with statistical tests.\n")

    # Assumption 3: Homogeneity of Variance Check - Levene's Test
    print("Assumption 3: Homogeneity of Variance Check (Levene's Test)")
    stat, p_value = stats.levene(*groups)
    print(f"Levene's Test statistic = {stat:.4f}, p-value = {p_value:.4f}")
    if p_value < 0.05:
        print("The variances are significantly different (p-value < 0.05). Homogeneity of variances assumption is violated.")
    else:
        print("The variances are not significantly different (p-value >= 0.05). Homogeneity of variances assumption is met.\n")
    
    # Visualize the data: Boxplot for homogeneity of variance and Q-Q plots for normality
    plt.figure(figsize=(12, 6))
    
    # Boxplot for homogeneity of variances
    plt.subplot(1, 2, 1)
    sns.boxplot(data=groups)
    plt.title("Boxplot for Homogeneity of Variances")
    
    # Q-Q plot for normality check
    plt.subplot(1, 2, 2)
    for i, group in enumerate(groups):
        stats.probplot(group, dist="norm", plot=plt)
        plt.title(f"Q-Q Plot for Group {i+1}")
        plt.show()

# Example usage: Simulating data for 3 groups
np.random.seed(42)

# Simulate data for 3 groups (Group 1, Group 2, Group 3)
group1 = np.random.normal(loc=100, scale=10, size=30)  # Mean = 100, SD = 10
group2 = np.random.normal(loc=105, scale=15, size=30)  # Mean = 105, SD = 15
group3 = np.random.normal(loc=110, scale=20, size=30)  # Mean = 110, SD = 20

# Check ANOVA assumptions for these groups
check_anova_assumptions([group1, group2, group3])
```

#### 24  Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results

```python
import numpy as np
from scipy.stats import chi2

# Step 1: Simulate a sample from a normal distribution
np.random.seed(42)  # For reproducibility

# Simulate data for a sample of 30 observations with a population variance of 225 (SD = 15)
sample_data = np.random.normal(loc=100, scale=15, size=30)

# Step 2: Calculate the sample variance
sample_variance = np.var(sample_data, ddof=1)  # Sample variance (ddof=1 for unbiased estimate)

# Step 3: Set the hypothesized population variance (H‚ÇÄ)
hypothesized_variance = 225  # Hypothesized population variance

# Step 4: Calculate the Chi-square test statistic
n = len(sample_data)  # Sample size
df = n - 1  # Degrees of freedom
chi_square_statistic = (df * sample_variance) / hypothesized_variance

# Step 5: Calculate the p-value for the Chi-square test statistic
p_value = 2 * min(chi2.cdf(chi_square_statistic, df), 1 - chi2.cdf(chi_square_statistic, df))

# Step 6: Print the results
print(f"Sample Variance: {sample_variance:.4f}")
print(f"Chi-square Statistic: {chi_square_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Step 7: Interpret the results
alpha = 0.05  # Significance level
if p_value < alpha:
    print("\nReject the null hypothesis: The population variance is significantly different from the hypothesized variance.")
else:
    print("\nFail to reject the null hypothesis: The population variance is not significantly different from the hypothesized variance.")
```

#### 25 Write a Python script to perform a Z-test for comparing proportions between two datasets or groups

```python
import numpy as np
from scipy.stats import norm

# Function to perform the Z-test for comparing proportions
def z_test_for_proportions(x1, n1, x2, n2):
    # Calculate sample proportions
    p1 = x1 / n1
    p2 = x2 / n2
    
    # Calculate pooled proportion
    P = (x1 + x2) / (n1 + n2)
    
    # Calculate the Z-statistic
    Z = (p1 - p2) / np.sqrt(P * (1 - P) * (1/n1 + 1/n2))
    
    # Calculate the p-value (two-tailed test)
    p_value = 2 * (1 - norm.cdf(abs(Z)))
    
    return Z, p_value

# Example usage
# Sample data: Group 1 (e.g., 60 successes out of 100 trials) and Group 2 (e.g., 50 successes out of 100 trials)
x1, n1 = 60, 100  # Group 1: 60 successes out of 100 trials
x2, n2 = 50, 100  # Group 2: 50 successes out of 100 trials

# Perform the Z-test
Z_statistic, p_value = z_test_for_proportions(x1, n1, x2, n2)

# Print the results
print(f"Z-statistic: {Z_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Step 2: Interpret the results
alpha = 0.05  # Significance level
if p_value < alpha:
    print("\nReject the null hypothesis: The proportions are significantly different.")
else:
    print("\nFail to reject the null hypothesis: The proportions are not significantly different.")
```
