#### 13 What is the Central Limit Theorem (CLT), and why is it critical in statistics

It states that the distribution of the sum (or average) of a large number of independent, identically distributed (i.i.d.) random variables approaches a normal distribution, no matter what the original distribution of the variables is, as long as the variables have a finite mean and variance.

#### 21  What are the assumptions for applying the Central Limit Theorem

The Central Limit Theorem (CLT) is a key concept in statistics that states that, regardless of the population's distribution, the sampling distribution of the sample mean will tend to be normal as the sample size becomes large enough. This allows statisticians to make inferences about population parameters using the normal distribution, even if the data itself isn‚Äôt normally distributed.

For the CLT to apply, certain assumptions must be met:

Random Sampling: The data must come from a random sample, ensuring it represents the population.

Independence: The sampled observations should be independent of each other. This means the selection of one sample doesn‚Äôt affect the others.

Sample Size: The sample size needs to be large enough. Generally, a sample size of 30 or more is considered large enough for the CLT to work, though smaller samples might be acceptable if the population is already normal.

Finite Variance: The population should have a finite variance (i.e., the spread of the population data should not be infinite).

#### 3 Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution

```python
def bernoulli_pdf(x, p):
    """
    Calculate the Probability Density Function (PDF) for a Bernoulli distribution.

    Parameters:
    - x: The outcome (0 or 1).
    - p: The probability of success (0 <= p <= 1).

    Returns:
    - The PDF value for the given x.
    """
    if x == 1:
        return p  # Probability of success
    elif x == 0:
        return 1 - p  # Probability of failure
    else:
        return 0  # Invalid value for a Bernoulli distribution (only 0 or 1 are valid)

# Example usage:
p = 0.7  # Probability of success
x_values = [0, 1]  # Possible outcomes in a Bernoulli distribution

for x in x_values:
    print(f"P(X = {x}) = {bernoulli_pdf(x, p)}")
```

#### 6 Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the discrete uniform distribution
low = 1      # Lower bound
high = 6     # Upper bound (inclusive)
size = 1000  # Number of samples

# Generate random samples from a discrete uniform distribution
samples = np.random.randint(low, high + 1, size)

# Calculate the CDF
values, counts = np.unique(samples, return_counts=True)
cdf = np.cumsum(counts) / size  # CDF is the cumulative sum of probabilities

# Plot the CDF
plt.step(values, cdf, where='post', color='skyblue', linewidth=2)
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.title('Cumulative Distribution Function (CDF) of a Discrete Uniform Distribution')
plt.xticks(values)
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()
```

#### 9 Write a Python function to calculate Z-scores from a dataset and plot them

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    """
    Calculate Z-scores for a given dataset.

    Parameters:
    - data: A list or numpy array of data points.

    Returns:
    - z_scores: A numpy array containing the Z-scores.
    """
    mean = np.mean(data)           # Calculate the mean of the data
    std_dev = np.std(data)         # Calculate the standard deviation of the data
    z_scores = (data - mean) / std_dev  # Z-score formula
    return z_scores

# Example dataset (e.g., test scores)
data = [50, 60, 65, 70, 85, 90, 95, 100, 110, 120]

# Calculate Z-scores
z_scores = calculate_z_scores(data)

# Plot the Z-scores
plt.plot(data, z_scores, marker='o', linestyle='-', color='skyblue', markersize=8)
plt.axhline(0, color='black',linewidth=1)  # Line at Z=0 for reference
plt.xlabel('Data Points')
plt.ylabel('Z-Scores')
plt.title('Z-Scores of the Dataset')
plt.grid(True)
plt.show()

# Print the Z-scores for reference
print(f"Z-scores: {z_scores}")
```

#### 10 Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the non-normal distribution (uniform distribution in this case)
low = 1      # Lower bound
high = 100   # Upper bound
size = 1000  # Number of samples to take
sample_size = 50  # Size of each sample

# Generate random samples from a non-normal distribution (Uniform distribution)
data = np.random.uniform(low, high, size)

# Function to simulate CLT by taking sample means
def apply_clt(data, sample_size, num_samples):
    sample_means = []
    for _ in range(num_samples):
        # Take a random sample of size 'sample_size' and compute its mean
        sample = np.random.choice(data, sample_size)
        sample_means.append(np.mean(sample))
    return sample_means

# Apply the CLT
num_samples = 1000
sample_means = apply_clt(data, sample_size, num_samples)

# Plotting the original distribution (Uniform) and the distribution of sample means
plt.figure(figsize=(12, 6))

# Plot the original data distribution (Uniform distribution)
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title('Original Non-Normal Distribution (Uniform)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True, axis='y', linestyle='--', alpha=0.7)

# Plot the distribution of sample means (CLT effect)
plt.subplot(1, 2, 2)
plt.hist(sample_means, bins=30, color='lightcoral', edgecolor='black', density=True)
plt.title('Distribution of Sample Means (CLT Applied)')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.grid(True, axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

def binomial_distribution(n, p, size):
    # Generate random variables (number of successes in n trials)
    random_variables = np.random.binomial(n, p, size)
    
    # Calculate the probability mass function (PMF) for each value of x (number of successes)
    x_values = np.arange(0, n+1)  # Possible outcomes (0 to n successes)
    pmf_values = binom.pmf(x_values, n, p)  # PMF for each number of successes
    
    return random_variables, x_values, pmf_values

# Parameters for the binomial distribution
n = 10   # Number of trials
p = 0.5  # Probability of success
size = 1000  # Number of samples to generate

# Generate random variables and calculate corresponding probabilities (PMF)
random_variables, x_values, pmf_values = binomial_distribution(n, p, size)

# Plot the histogram of the generated random variables (number of successes)
plt.figure(figsize=(12, 6))

# Plot 1: Histogram of random variables (Binomial distribution)
plt.subplot(1, 2, 1)
plt.hist(random_variables, bins=np.arange(n+2)-0.5, edgecolor='black', density=True, color='skyblue')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title(f'Histogram of Binomial Distribution (n={n}, p={p})')
plt.xticks(np.arange(0, n+1))
plt.grid(True, axis='y', linestyle='--', alpha=0.7)

# Plot 2: Probability Mass Function (PMF) of the binomial distribution
plt.subplot(1, 2, 2)
plt.stem(x_values, pmf_values, basefmt=" ", linefmt='r-', markerfmt='ro')  # Removed 'use_line_collection'
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title(f'Probability Mass Function (PMF) of Binomial Distribution (n={n}, p={p})')
plt.xticks(np.arange(0, n+1))
plt.grid(True, axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

#### 14 Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(data_point, mean, std_dev):
    """
    Calculate the Z-score for a given data point.
    
    Parameters:
    - data_point: The data point whose Z-score is to be calculated.
    - mean: The mean of the distribution.
    - std_dev: The standard deviation of the distribution.
    
    Returns:
    - Z-score for the given data point.
    """
    z_score = (data_point - mean) / std_dev
    return z_score

def plot_standard_normal_distribution(data_point, z_score, mean, std_dev):
    """
    Plot the standard normal distribution and highlight the given data point.
    
    Parameters:
    - data_point: The data point to be compared with the standard normal distribution.
    - z_score: The Z-score of the data point.
    - mean: The mean of the normal distribution.
    - std_dev: The standard deviation of the normal distribution.
    """
    # Create x values for plotting the standard normal distribution
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x, mean, std_dev)  # PDF of the standard normal distribution
    
    # Plot the standard normal distribution
    plt.plot(x, y, label="Standard Normal Distribution", color='skyblue', linewidth=2)
    
    # Highlight the given data point
    plt.axvline(x=data_point, color='red', linestyle='--', label=f"Data Point (Z = {z_score:.2f})")
    plt.fill_between(x, y, where=(x >= data_point), color='red', alpha=0.3)
    
    # Adding labels and title
    plt.xlabel('X')
    plt.ylabel('Probability Density')
    plt.title('Standard Normal Distribution with Highlighted Data Point')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage:
mean = 100  # Mean of the distribution
std_dev = 15  # Standard deviation of the distribution
data_point = 120  # The given data point

# Calculate the Z-score for the data point
z_score = calculate_z_score(data_point, mean, std_dev)

# Plot the standard normal distribution and compare it with the data point
plot_standard_normal_distribution(data_point, z_score, mean, std_dev)

# Print the Z-score for reference
print(f"Z-score for data point {data_point}: {z_score:.2f}")
```

#### 15  Implement hypothesis testing using Z-statistics for a sample dataset

```python
import numpy as np
from scipy.stats import norm

# Sample data (test scores from a class)
sample_data = [78, 82, 85, 90, 88, 76, 79, 95, 83, 81]

# Parameters for the hypothesis test
population_mean = 75     # Known population mean
population_std = 10      # Known population standard deviation
sample_size = len(sample_data)

# Calculate the sample mean
sample_mean = np.mean(sample_data)

# Z-statistic formula
z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Calculate the p-value (two-tailed test)
p_value = 2 * (1 - norm.cdf(abs(z_statistic)))  # Two-tailed test

# Significance level (alpha)
alpha = 0.05

# Output the results
print(f"Sample Mean: {sample_mean}")
print(f"Z-statistic: {z_statistic:.2f}")
print(f"P-value: {p_value:.4f}")

# Decision: reject the null hypothesis if p-value < alpha
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference.")
```

#### 16 Create a confidence interval for a dataset using Python and interpret the result

```python
import numpy as np
import scipy.stats as stats

def confidence_interval(data, confidence_level=0.95):
    """
    Calculate the confidence interval for the mean of a dataset.

    Parameters:
    - data: List or numpy array of data points
    - confidence_level: The desired confidence level (default is 95%)

    Returns:
    - A tuple containing the lower and upper bounds of the confidence interval.
    """
    # Sample statistics
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # Sample standard deviation (ddof=1 for unbiased estimator)
    sample_size = len(data)
    
    # Degrees of freedom
    df = sample_size - 1
    
    # t-score for the desired confidence level
    t_score = stats.t.ppf((1 + confidence_level) / 2, df)
    
    # Margin of error
    margin_of_error = t_score * (sample_std / np.sqrt(sample_size))
    
    # Confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error
    
    return lower_bound, upper_bound

# Example dataset (e.g., sample test scores)
data = [78, 82, 85, 90, 88, 76, 79, 95, 83, 81]

# Calculate the 95% confidence interval
lower, upper = confidence_interval(data, confidence_level=0.95)

# Print the result
print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")
```

#### 17 Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean

```python
import numpy as np
import scipy.stats as stats

def confidence_interval(data, confidence_level=0.95):
    """
    Calculate the confidence interval for the mean of a dataset.

    Parameters:
    - data: List or numpy array of data points
    - confidence_level: The desired confidence level (default is 95%)

    Returns:
    - A tuple containing the lower and upper bounds of the confidence interval.
    """
    # Sample statistics
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # Sample standard deviation (ddof=1 for unbiased estimator)
    sample_size = len(data)
    
    # Degrees of freedom
    df = sample_size - 1
    
    # t-score for the desired confidence level
    t_score = stats.t.ppf((1 + confidence_level) / 2, df)
    
    # Margin of error
    margin_of_error = t_score * (sample_std / np.sqrt(sample_size))
    
    # Confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error
    
    return lower_bound, upper_bound

# Generate random data from a normal distribution
np.random.seed(0)  # For reproducibility
mean = 100          # True mean of the distribution
std_dev = 15        # True standard deviation
sample_size = 50    # Sample size

# Generate random data from a normal distribution
data = np.random.normal(mean, std_dev, sample_size)

# Calculate the 95% confidence interval
lower, upper = confidence_interval(data, confidence_level=0.95)

# Print the result
print(f"Sample Mean: {np.mean(data):.2f}")
print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_normal_pdf(mean, std_dev):
    # Generate a range of x values (from -4 to 4 standard deviations)
    x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)
    
    # Calculate the PDF for the normal distribution
    pdf = norm.pdf(x, mean, std_dev)
    
    # Plot the PDF
    plt.plot(x, pdf, label=f'Normal Distribution (mean={mean}, std={std_dev})', color='blue', linewidth=2)
    
    # Add labels and title
    plt.xlabel('X')
    plt.ylabel('Probability Density')
    plt.title('Probability Density Function (PDF) of Normal Distribution')
    
    # Add grid and legend
    plt.grid(True)
    plt.legend()
    
    # Show the plot
    plt.show()

# Example usage:
mean = 0         # Mean of the distribution
std_dev = 1      # Standard deviation of the distribution

# Call the function to plot the normal distribution's PDF
plot_normal_pdf(mean, std_dev)
```

#### 19 5 Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

def plot_poisson_cdf(lmbda, max_x):
    # Calculate the CDF for each value from 0 to max_x
    x_values = np.arange(0, max_x + 1)
    cdf_values = poisson.cdf(x_values, lmbda)
    
    # Plot the CDF
    plt.step(x_values, cdf_values, where='post', color='skyblue', linewidth=2, label=f'Poisson CDF (Œª={lmbda})')
    plt.xlabel('Number of Events (k)')
    plt.ylabel('Cumulative Probability')
    plt.title(f'Cumulative Distribution Function (CDF) of Poisson Distribution (Œª={lmbda})')
    plt.grid(True)
    plt.legend()
    plt.show()

# Example usage:
lambda_value = 5  # Rate (mean number of occurrences)
max_x = 15  # Maximum value for which to calculate the CDF

# Call the function to plot the Poisson CDF
plot_poisson_cdf(lambda_value, max_x)
```

#### 20 5 Simulate a random variable using a continuous uniform distribution and calculate its expected value

```python
import numpy as np

def simulate_uniform_random_variable(a, b, size):
    """
    Simulate a random variable from a continuous uniform distribution.

    Parameters:
    - a: Lower bound of the distribution
    - b: Upper bound of the distribution
    - size: Number of random samples to generate

    Returns:
    - random_samples: Array of random samples from the continuous uniform distribution
    """
    random_samples = np.random.uniform(a, b, size)
    return random_samples

def calculate_expected_value(a, b):
    """
    Calculate the expected value for a continuous uniform distribution.

    Parameters:
    - a: Lower bound of the distribution
    - b: Upper bound of the distribution

    Returns:
    - Expected value of the distribution
    """
    return (a + b) / 2

# Example usage
a = 0  # Lower bound
b = 10  # Upper bound
size = 10000  # Number of random samples to generate

# Simulate random variable from a continuous uniform distribution
random_samples = simulate_uniform_random_variable(a, b, size)

# Calculate the expected value
expected_value = calculate_expected_value(a, b)

# Display the results
print(f"Expected Value of the Continuous Uniform Distribution: {expected_value}")
print(f"Sample Mean (Average of simulated values): {np.mean(random_samples)}")
```

#### 21 Write a Python program to compare the standard deviations of two datasets and visualize the difference

```python
import numpy as np
import matplotlib.pyplot as plt

def compare_standard_deviations(data1, data2):
    """
    Compare the standard deviations of two datasets and visualize the difference.
    
    Parameters:
    - data1: First dataset (list or numpy array)
    - data2: Second dataset (list or numpy array)
    
    Returns:
    - standard_deviation1: Standard deviation of the first dataset
    - standard_deviation2: Standard deviation of the second dataset
    """
    # Calculate the standard deviation for each dataset
    std_dev1 = np.std(data1, ddof=1)  # Sample standard deviation (ddof=1)
    std_dev2 = np.std(data2, ddof=1)  # Sample standard deviation (ddof=1)
    
    # Plotting the standard deviations using a bar plot
    plt.bar(['Dataset 1', 'Dataset 2'], [std_dev1, std_dev2], color=['skyblue', 'lightcoral'])
    plt.xlabel('Datasets')
    plt.ylabel('Standard Deviation')
    plt.title('Comparison of Standard Deviations of Two Datasets')
    plt.show()
    
    return std_dev1, std_dev2

# Example datasets
dataset1 = [12, 15, 18, 20, 24, 30, 35, 40, 45, 50]  # Dataset 1
dataset2 = [100, 105, 110, 120, 130, 135, 140, 150, 160, 170]  # Dataset 2

# Compare and visualize the standard deviations
std_dev1, std_dev2 = compare_standard_deviations(dataset1, dataset2)

# Print the standard deviations for reference
print(f"Standard Deviation of Dataset 1: {std_dev1:.2f}")
print(f"Standard Deviation of Dataset 2: {std_dev2:.2f}")
```

#### 22 Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution

```python
import numpy as np

def calculate_range_and_iqr(data):
    """
    Calculate the range and interquartile range (IQR) for a dataset.
    
    Parameters:
    - data: A list or numpy array of data points.
    
    Returns:
    - range_value: The range of the dataset.
    - iqr_value: The interquartile range (IQR) of the dataset.
    """
    # Calculate the range
    range_value = np.max(data) - np.min(data)
    
    # Calculate the IQR (Q3 - Q1)
    Q1 = np.percentile(data, 25)  # First quartile (25th percentile)
    Q3 = np.percentile(data, 75)  # Third quartile (75th percentile)
    iqr_value = Q3 - Q1
    
    return range_value, iqr_value

# Generate a dataset from a normal distribution
np.random.seed(0)  # For reproducibility
mean = 50          # Mean of the normal distribution
std_dev = 10       # Standard deviation of the normal distribution
sample_size = 1000 # Number of data points to generate

# Generate random data from a normal distribution
data = np.random.normal(mean, std_dev, sample_size)

# Calculate the range and IQR of the generated data
range_value, iqr_value = calculate_range_and_iqr(data)

# Print the results
print(f"Range of the dataset: {range_value:.2f}")
print(f"Interquartile Range (IQR) of the dataset: {iqr_value:.2f}")
```

#### 23 5 Implement Z-score normalization on a dataset and visualize its transformation

```python
import numpy as np
import matplotlib.pyplot as plt

def z_score_normalization(data):
    """
    Normalize the dataset using Z-score normalization.
    
    Parameters:
    - data: A list or numpy array of data points.
    
    Returns:
    - normalized_data: The Z-score normalized version of the data.
    """
    mean = np.mean(data)
    std_dev = np.std(data, ddof=0)  # Population standard deviation (ddof=0)
    
    normalized_data = (data - mean) / std_dev
    return normalized_data

# Generate a sample dataset from a normal distribution
np.random.seed(0)  # For reproducibility
original_data = np.random.normal(100, 15, 1000)  # Mean=100, Std=15, Sample size=1000

# Apply Z-score normalization to the data
normalized_data = z_score_normalization(original_data)

# Plot the original and normalized data for comparison
plt.figure(figsize=(12, 6))

# Plot the histogram of the original data
plt.subplot(1, 2, 1)
plt.hist(original_data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Original Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)

# Plot the histogram of the normalized data
plt.subplot(1, 2, 2)
plt.hist(normalized_data, bins=30, color='lightcoral', edgecolor='black', alpha=0.7)
plt.title('Z-Score Normalized Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)

plt.tight_layout()
plt.show()
```

#### 24 Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution

```python
import numpy as np
from scipy.stats import skew, kurtosis

def calculate_skewness_kurtosis(data):
    """
    Calculate the skewness and kurtosis of a dataset.
    
    Parameters:
    - data: A list or numpy array of data points.
    
    Returns:
    - skewness: The skewness of the dataset.
    - kurtosis: The kurtosis of the dataset.
    """
    # Calculate skewness
    skew_value = skew(data)
    
    # Calculate kurtosis (excess kurtosis)
    kurtosis_value = kurtosis(data)
    
    return skew_value, kurtosis_value

# Generate a dataset from a normal distribution
np.random.seed(0)  # For reproducibility
mean = 0           # Mean of the normal distribution
std_dev = 1        # Standard deviation of the normal distribution
sample_size = 1000 # Number of data points to generate

# Generate random data from a normal distribution
data = np.random.normal(mean, std_dev, sample_size)

# Calculate skewness and kurtosis
skew_value, kurtosis_value = calculate_skewness_kurtosis(data)

# Print the results
print(f"Skewness of the dataset: {skew_value:.4f}")
print(f"Kurtosis of the dataset (Excess Kurtosis): {kurtosis_value:.4f}")
```
